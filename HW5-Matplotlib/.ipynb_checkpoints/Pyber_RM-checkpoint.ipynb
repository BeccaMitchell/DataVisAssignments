```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

city_data_load = "data/city_data.csv"
ride_data_load = "data/ride_data.csv"

city_data = pd.read_csv(city_data_load)
ride_data = pd.read_csv(ride_data_load)

city_ride_data = pd.merge(ride_data, city_data, how="left", on=["city", "city"])
city_ride_data.head()

# Display the data table for preview
```

```python
# Obtain the x and y coordinates for each of the three city types
ride_count = city_ride_data.groupby(["city", "type", "driver_count"])["ride_id"].count()
avg_fare = city_ride_data.groupby(["city", "type", "driver_count"])["fare"].mean()

bubble_df = pd.DataFrame({
    "Average Fare" : avg_fare,
    "Total Rides" : ride_count
}).reset_index()

urban =plt.scatter(x=bubble_df[bubble_df["type"]=="Urban"]["Total Rides"], y=bubble_df[bubble_df["type"]=="Urban"]["Average Fare"], 
                   label="Urban", s=(bubble_df[bubble_df["type"]=="Urban"]["driver_count"])*15, color="coral", edgecolors="black", alpha=.70)
suburban =plt.scatter(x=bubble_df[bubble_df["type"]=="Suburban"]["Total Rides"], y=bubble_df[bubble_df["type"]=="Suburban"]["Average Fare"], 
                      label="Suburban", s=(bubble_df[bubble_df["type"]=="Suburban"]["driver_count"])*15, color="lightskyblue", edgecolors="black", alpha=.70)
rural =plt.scatter(x=bubble_df[bubble_df["type"]=="Rural"]["Total Rides"], y=bubble_df[bubble_df["type"]=="Rural"]["Average Fare"],
                   label="Rural", s=(bubble_df[bubble_df["type"]=="Rural"]["driver_count"])*15, color="gold", edgecolors="black", alpha=.70 )

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid(True)
plt.text(43, 35,"Note: \nCircle size correlates with \ndriver count per city")

lgnd=plt.legend([urban, suburban, rural],['Urban', 'Suburban', 'Rural'], title="City Types", shadow=True)
lgnd.legendHandles[0]._sizes=[75]
lgnd.legendHandles[1]._sizes=[75]
lgnd.legendHandles[2]._sizes=[75]

plt.savefig("Pyber/Pyber_Ride_Sharing_2016.png")
plt.show()
```

```python
#% of Total Fares by City Type
city_type_fare = city_ride_data.groupby("type")["fare"].sum()
total_fares = city_type_fare.sum()
pct_fare_type = (city_type_fare/total_fares)*100

colors = ["gold", "lightskyblue", "lightcoral"]
labels = ["Rural", "Suburban", "Urban"]
angle = 120
expl = (0, 0, 0.1)

plt.pie(pct_fare_type, colors=colors, labels=labels, shadow=True, startangle = angle, explode = expl, autopct='%1.1f%%')
plt.title("% of Total Fares by City Type")

plt.savefig("Pyber/Pyber_TotalFaresbyCityType.png")
plt.show()
```

```python
#% of Total Rides by City Type
city_type_rides = city_ride_data.groupby("type")["ride_id"].count()
total_rides = city_type_rides.sum()
pct_rides_type = (city_type_rides/total_rides)*100

plt.pie(pct_rides_type, colors=colors, labels=labels, shadow=True, startangle = angle, explode = expl, autopct='%1.1f%%')
plt.title("% of Total Rides by City Type")

plt.savefig("Pyber/Pyber_TotalRidesbyCityType.png")
plt.show()
```

```python
#% of Total Drivers by City Type
city_type_drivers = city_data.groupby("type")["driver_count"].sum()
total_drivers = city_type_drivers.sum()
pct_drivers_type = (city_type_drivers/total_drivers)*100

plt.pie(pct_drivers_type, colors=colors, labels=labels, shadow=True, startangle = angle, explode = expl, autopct='%1.1f%%')
plt.title("% of Total Drivers by City Type")

plt.savefig("Pyber/Pyber_TotalDriversbyCityType.png")
plt.show()
```
